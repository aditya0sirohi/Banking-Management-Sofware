```python
import mysql.connector
import re
import sqlite3
import random  # Import the random module
from datetime import datetime, timedelta
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='Aditya@2004',
    database='minibank'
)
cursor = connection.cursor()
def random_accno():
    return f"{random.randint(100000, 999999)}"

class User():
    def __init__(self, account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey):
        self.account_number = account_number or random_accno()
        self.customer_name = customer_name
        self.customer_address = customer_address
        self.customer_contact = customer_contact
        self.customer_age = customer_age
        self.customer_gender = customer_gender
        self.balance = balance
        self.passkey = passkey

    def show_details(self):
        print("Personal Details")
        print("Account number:", self.account_number)
        print("Customer Name:", self.customer_name)
        print("Customer Contact:", self.customer_contact)
        print("Customer Age:", self.customer_age)
        print("Customer Address:", self.customer_address)
        print("Customer Gender:", self.customer_gender)
        print("Account Balance:", self.balance)

    def validate_passkey(self):
        # Validate if passkey is strong enough
        if (
            re.search(r'[0-9]', self.passkey) and
            re.search(r'[A-Z]', self.passkey) and
            re.search(r'[@#$]', self.passkey) and
            len(self.passkey) >= 8
        ):
            return True
        else:
            print("Passkey is not strong enough. It should contain at least one digit, one uppercase letter, one special character (@, #, $), and be at least 8 characters long.")
            return False

    def authenticate(self, entered_passkey):
        # Check if entered passkey matches the stored passkey
        return self.passkey == entered_passkey
    def save_to_database(self):
        insert_user_query = """
        INSERT INTO users (account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)
        VALUES (%s, %s, %s, %s, %s, %s, %s, %s)
        """
        user_data = (self.account_number, self.customer_name, self.customer_contact, self.customer_age, self.customer_address, self.customer_gender, self.balance, self.passkey)
        cursor.execute(insert_user_query, user_data)
        connection.commit()

        cursor.close()
        connection.close()
        
class Bank(User):
    def __init__(self, account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey):
        super().__init__(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        self.balance -= amount

    def get_balance(self):
        return self.balance

    def charges(self):
        print("Net banking services are free for our customers")
        total_charge = 0
        c = input("Do you have a debit card? (y/n): ")
        if c.lower() == 'y':
            total_charge += 10
        else:
            print("User doesn't own a debit card")

        d = input("Did you opt for text messaging services? (y/n): ")
        if d.lower() == 'y':
            total_charge += 5
        else:
            print("User didn't opt for messaging services")

        # Deduct charges from the balance
        self.balance -= total_charge
        print("Total charges deducted:", total_charge)

class CurrentAccount(Bank):
    def __init__(self, account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey):
        super().__init__(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)

    def withdraw(self, amount):
        if self.balance - amount >= -50000:
            self.balance -= amount
        else:
            print("Insufficient balance!")

    def calculate_interest(self):
        if self.balance > 0:
            interest = self.balance * 0.04
            self.balance += interest

    def apply_surcharge(self):
        if self.balance < 0:
            surcharge = self.balance * 0.06
            self.balance += surcharge

    def monthly_balance_sheet(self):
        print("Monthly Balance Sheet")
        super().show_details()
        print("Account Balance:", self.balance)

class SavingsAccount(Bank):
    def __init__(self, account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey):
        super().__init__(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)

    def deposit(self, amount):
        if amount >= 100:
            self.balance += amount
        else:
            print("Minimum deposit amount is RM 100!")

    def withdraw(self, amount):
        if self.balance - amount >= 50:
            self.balance -= amount
        else:
            print("Insufficient balance!")

    def calculate_interest(self):
        interest = self.balance * 0.05 / 12
        self.balance += interest

    def apply_fine(self):
        if self.balance < 50:
            fine = self.balance * 0.1
            self.balance -= fine

    def monthly_balance_sheet(self):
        print("Monthly Balance Sheet")
        super().show_details()
        print("Account Balance:", self.balance)

def create_account(account_number, customer_name, balance, customer_contact, customer_age, customer_gender, customer_address, passkey):
    account_type = input("Enter your type of account (current/savings): ")
    if account_type.lower() == "current":
        return CurrentAccount(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)
    elif account_type.lower() == "savings":
        return SavingsAccount(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)
    else:
        raise ValueError("Invalid account type")
import math
class Loan:
    def __init__(self, user, amount, rate_of_interest, max_duration):
        self.user = user
        self.amount = amount
        self.rate_of_interest = rate_of_interest
        self.max_duration = max_duration

    def check_eligibility(self, min_balance_required):
        return self.user.balance >= min_balance_required

    def calculate_interest(self, duration):
        compound_interest = self.amount * (1 + self.rate_of_interest / 100) ** duration
        return compound_interest - self.amount

class CarLoan(Loan):
    def __init__(self, user):
        super().__init__(user, 0, 6, 60)

    def check_eligibility(self):
        min_balance_required = 20000
        return super().check_eligibility(min_balance_required)

    def apply_loan(self, loan_amount):
        if self.check_eligibility():
            self.amount = loan_amount
            print("Car Loan approved!")

    def submit_instalment(self, duration):
        # Assuming duration is in months
        if duration > 2:
            self.rate_of_interest = 8
            print("Interest rate increased due to missed instalments.")

class TwoWheelerLoan(Loan):
    def __init__(self, user):
        super().__init__(user, 0, 4, 60)

    def check_eligibility(self):
        min_balance_required = 5000
        return super().check_eligibility(min_balance_required)

    def apply_loan(self, loan_amount):
        if self.check_eligibility():
            self.amount = loan_amount
            print("Two Wheeler Loan approved!")

    def submit_instalment(self, duration):
        # Assuming duration is in months
        if duration > 2:
            self.rate_of_interest = 6
            print("Interest rate increased due to missed instalments.")

class HomeLoan(Loan):
    def __init__(self, user):
        super().__init__(user, 0, 8, 300)

    def check_eligibility(self):
        average_balance_required = 50000  # Average balance over the last 2 years
        return self.user.balance >= average_balance_required

    def apply_loan(self, loan_amount):
        if self.check_eligibility():
            self.amount = loan_amount
            print("Home Loan approved!")

class MaintenanceLoan(Loan):
    def __init__(self, user):
        super().__init__(user, 0, 5, 60)

    def apply_loan(self, loan_amount):
        print("Maintenance Loan approved!")


class PersonalLoan(Loan):
    def __init__(self, user):
        super().__init__(user, 0, 10, 60)

    def apply_loan(self, loan_amount):
        # Assuming loan amount is up to 20% of the annual salary
        if loan_amount <= 0.2 * self.user.balance:
            self.amount = loan_amount
            print("Personal Loan approved!")
def main():
    account = None
    while True:
        print("\n___MENU___")
        print("1. Create account")
        print("2. Check account information")
        print("3. Deposit money")
        print("4. Withdraw money")
        print("5. Check if account type conditions met?")
        print("6. Monthly balance sheet")
        print("7. Check charges for net banking")
        print("8. Apply loan")
        print("9. Apply home loan")
        print("10. Apply personal loan")
        print("11.Exit")

        choice = int(input("Enter your choice (1-11): "))

        if choice == 1:
            passkey = input("Enter your password:")
            account_number = random_accno()
            customer_name = input("Enter customer name: ")
            balance = int(input("Enter initial balance: "))
            customer_contact = input("Enter contact number: ")
            customer_age = int(input("Enter your age: "))
            customer_gender = input("Gender(M/F/O): ")
            customer_address = input("Enter address: ")

            account = create_account(account_number, customer_name, balance, customer_contact, customer_age, customer_gender, customer_address, passkey)

            user = User(account_number, customer_name, customer_contact, customer_age, customer_address, customer_gender, balance, passkey)

        elif choice == 2:
             if user.validate_passkey():
                 entered_passkey = input("Enter your passkey again for authentication: ")
                 if user.authenticate(entered_passkey):
                     print("Authentication successful.")
                     if account:
                         account.show_details()
                     else:
                         print("No account created yet or authentication failed")

        elif choice == 3:
            if account:
                amount = int(input("\nEnter deposit amount: "))
                account.deposit(amount)
                print(f"Deposited RM {amount}. Updated Balance: RM {account.get_balance()}")
            else:
                print("No account created yet.")

        elif choice == 4:
            if account:
                amount = int(input("Enter withdraw amount: "))
                account.withdraw(amount)
                print(f"Withdrew RM {amount}. Updated Balance: RM {account.get_balance()}")
            else:
                print("No account created yet.")

        elif choice == 5:
            if account:
                if isinstance(account, CurrentAccount):
                    amount = int(input("Enter deposit amount for current account: "))
                    account.deposit(amount)
                    print(f"Deposited RM {amount}. Updated Balance: RM {account.get_balance()}")

                    amount = int(input("Enter withdraw amount for current account: "))
                    account.withdraw(amount)
                    print(f"Withdrew RM {amount}. Updated Balance: RM {account.get_balance()}")

                    interest = int(input("Enter the interest for current account: "))
                    account.calculate_interest()
                    print(f"Total interest applied RM {interest}. Updated Balance: RM {account.get_balance()}")

                    surcharge = int(input("Enter the surcharge: "))
                    account.apply_surcharge()
                    print(f"Applied surcharge RM {surcharge}. Updated Balance: RM {account.get_balance()}")

                elif isinstance(account, SavingsAccount):
                    amount = int(input("Enter deposit amount for savings account: "))
                    account.deposit(amount)
                    print(f"Deposited RM {amount}. Updated Balance: RM {account.get_balance()}")

                    amount = int(input("Enter withdraw amount for savings account: "))
                    account.withdraw(amount)
                    print(f"Withdrew RM {amount}. Updated Balance: RM {account.get_balance()}")

                    interest = int(input("Enter the interest for savings account: "))
                    account.calculate_interest()
                    print(f"Total interest applied RM {interest}. Updated Balance: RM {account.get_balance()}")

                    fine = int(input("Enter the fine: "))
                    account.apply_fine()
                    print(f"Applied fine RM {fine}. Updated Balance: RM {account.get_balance()}")

                else:
                    raise ValueError("Invalid account type")
            else:
                print("No account created yet.")

        elif choice == 6:
            if account:
                account.monthly_balance_sheet()
            else:
                print("No account created yet.")

        elif choice == 7:
            if account:
                account.charges()
            else:
                print("No account created yet.")

        elif choice == 8:
            if account:
                account.save_to_database()
                loan_amount = int(input("Enter loan amount: "))
                car_loan = CarLoan(user)
                if car_loan.apply_loan(loan_amount):
                    # For demonstration purposes, assume duration is 12 months
                    car_loan.submit_instalment(12)
            else:
                print("No account created yet.")

        elif choice == 9:
            if account:
                account.save_to_database()
                loan_amount = int(input("Enter loan amount: "))
                home_loan = HomeLoan(user)
                if home_loan.apply_loan(loan_amount):
                    # For demonstration purposes, assume duration is 12 months
                    home_loan.submit_instalment(12)
            else:
                print("No account created yet.")

        elif choice == 10:
            if account:
                account.save_to_database()
                loan_amount = int(input("Enter loan amount: "))
                personal_loan = PersonalLoan(user)
                if personal_loan.apply_loan(loan_amount):
                    # For demonstration purposes, assume duration is 12 months
                    personal_loan.submit_instalment(12)
            else:
                print("No account created yet.")

        elif choice == 11:
            print("Exiting program. Goodbye!")
            break
            

        else:
            print("Invalid Choice")

if __name__ == "__main__":
    main()
```
